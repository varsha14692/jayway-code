```python
import pandas as pd

# importing the dataset

df = pd.read_csv('ecommerce_data.csv')

# Remove duplicate rows

remdup = df.drop_duplicates(keep='first')
# remove the rows if quantity is missing OR convert/replace the missing value (either replace quantity as '1' or use median function)

ecod = remdup.dropna(subset=['Quantity'],how='all')
# OR 

# ecod = remdup['Quantity'].fillna(value='1', inplace=True)
# ecod = remdup['Quantity'].fillna(value=remdup['Quantity'].median(), inplace=True)

# add new column : TotalSales = Quantity * UnitPrice

TotalSales = ecod['Quantity'] * ecod['UnitPrice']
ecodu = pd.DataFrame(df)
ecodu['TotalSales'] = TotalSales

#ecod['TotalSales'] = TotalSales

# Aggregate the data : Max(TotalSales) group by country, Estore_id order by TotalSales asc/desc

fd = ecodu.groupby(['Country','Estore_id'], sort=True)['TotalSales'].mean()

ecodu.sort_values(by='TotalSales', ascending=True)
```
